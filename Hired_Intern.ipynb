```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Applications.csv")
df
```

```python
df.isnull().sum()
```

```python
columns_to_drop = ['Name','Degree','Other skills','Stream','Current Year Of Graduation','Performance_ PG','Performance_ UG','Performance_ 12','Performance_ 10']
df = df.drop(columns_to_drop, axis=1)
```

```python
df
```

```python
df = df.drop('Are you available for 3 months, starting immediately, for a full-time work from home internship?', axis=1)
```

```python
df
```

```python
df.describe()
```

```python
def is_hired(row):
    # Criteria 1: If intern has a minimum of 2 data points for all features, they are hired
    if row['Python (out of 3)'] >= 2 and row['Machine Learning (out of 3)'] >= 2 and row['Natural Language Processing\n(NLP) (out of\n3)'] >= 2 and row['Deep Learning (out of 3)'] >= 2:
        return 1  # Hired

    # Criteria 2: If intern knows Machine Learning and either Deep Learning or NLP, they are hired
    if row['Machine Learning (out of 3)'] >= 2 and (row['Deep Learning (out of 3)'] >= 1 or row['Natural Language Processing\n(NLP) (out of\n3)'] >= 1):
        return 1  # Hired

    # Criteria 3: Machine Learning is mandatory
    if row['Machine Learning (out of 3)'] >= 2 and row['Python (out of 3)'] >= 2:
        return 1  # Hired

    return 0  # Not Hired

df['Hired_Or_Not'] = df.apply(is_hired, axis=1)
```

```python
df
```

```python
df['Hired_Or_Not'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('Hired_Or_Not',axis=1)
y = df['Hired_Or_Not']

X_train,X_test,y_train,y_test = train_test_split(X , y , test_size=0.3)
```

```python
len(X_test)
```

```python
len(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression()
train_data=classifier.fit(X_train, y_train)
```

```python
y_pred = train_data.predict(X_test)
```
